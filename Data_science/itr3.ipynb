```python
import matplotlib.pyplot as plt
import pandas as pd

raw_data=pd.read_csv("data.csv", header=0, sep=",")

raw_data.dropna(axis=0, inplace=True)

raw_data['Average_Pulse']=raw_data['Average_Pulse'].astype(float)
raw_data['Max_Pulse']=raw_data['Max_Pulse'].astype(float)
print(raw_data)

raw_data.plot(x ='Average_Pulse', y='Calorie_Burnage', kind='line'),
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()
```
