```python
import datetime

# Текущая дата и время
now = datetime.datetime.now()
print(now)

# Создание объекта datetime
dt = datetime.datetime(2023, 10, 26, 10, 30, 0)
print(dt)

# Доступ к компонентам даты и времени
print(dt.year)
print(dt.month)
print(dt.day)
print(dt.hour)
print(dt.minute)
print(dt.second)

# Разница во времени (timedelta)
future_dt = dt + datetime.timedelta(days=7, hours=3)
print(future_dt)

past_dt = dt - datetime.timedelta(weeks=2)
print(past_dt)

time_diff = future_dt - dt
print(time_diff)
print(time_diff.total_seconds())

# Форматирование даты и времени
formatted_dt = dt.strftime("%Y-%m-%d %H:%M:%S")
print(formatted_dt)

# Парсинг строки в datetime
date_string = "2024-05-15 14:00:00"
parsed_dt = datetime.datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S")
print(parsed_dt)

# Только дата
today = datetime.date.today()
print(today)

date_obj = datetime.date(2025, 1, 1)
print(date_obj)

# Только время
current_time = datetime.datetime.now().time()
print(current_time)

time_obj = datetime.time(17, 45, 30)
print(time_obj)

# Сравнение datetime объектов
dt1 = datetime.datetime(2023, 1, 1)
dt2 = datetime.datetime(2023, 1, 2)
print(dt1 < dt2)
print(dt1 == dt2)

# Замена компонентов
new_dt = dt.replace(year=2024, month=11)
print(new_dt)
```
