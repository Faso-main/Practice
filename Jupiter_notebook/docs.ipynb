**Общее назначение:**  
Программа сравнивает производительность однопоточных и многопоточных методов численного интегрирования (прямоугольников, трапеций, Монте-Карло) и анализирует ускорение.

---

**Архитектура:**

1. **Интерфейс** (Windows Forms):
   - Поля ввода: `a`, `b`, `n` (пределы и число разбиений), `threads` (число потоков)
   - Кнопки: Single (однопоточный расчёт), Multi (многопоточный), Clear (очистка)
   - Таблица `dataResults` для вывода результатов

2. **Методы интегрирования**:
   - `SingleRectangle()` / `MultiRectangle()` — метод прямоугольников
   - `SingleTrapezoid()` / `MultiTrapezoid()` — метод трапеций  
   - `SingleMonteKarlo()` / `MultiMonteKarlo()` — метод Монте-Карло

3. **Параллелизм**:
   - `Parallel.For` с ограничением потоков (`MaxDegreeOfParallelism`)
   - Локальные переменные потоков для избежания конфликтов
   - `lock` для безопасного суммирования результатов

4. **Анализ производительности**:
   - Замер времени через `Stopwatch`
   - Расчёт экспериментального ускорения (отношение времени single/multi)
   - Теоретическое ускорение по законам Амдала и Густафсона

---

**Ключевые особенности:**
- Сравнивает три разных численных метода
- Измеряет реальное и теоретическое ускорение
- Thread-safe реализация многопоточных версий
- Визуализация результатов в таблице

**Проблемы/Улучшения:**
- Жёстко зафиксированная доля последовательной части (S=0.01) в законах Амдала/Густафсона
- Нет обработки ошибок в многопоточных методах
- Возможна оптимизация без использования `lock` через более эффективные методы агрегации
